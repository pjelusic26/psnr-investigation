```python
import numpy as np
import skimage
import skimage.measure as msr
import matplotlib.pyplot as plt
from matplotlib.pyplot import imshow
from matplotlib import rcParams
from PIL import Image
```

```python
# Read images
imgZit0 = Image.open("psnrEdit-1x.tif")
imgZit1 = Image.open("psnrOrig.tif")
imgZit2 = Image.open("psnrEdit-2x.tif")
imgZit4 = Image.open("psnrEdit-4x.tif")

# figure size in inches optional
rcParams['figure.figsize'] = 18 ,18

# Display images side by side
fig, ax = plt.subplots(2,2)
ax[0, 0].imshow(imgZit0);
ax[0, 0].set_title('No Zit')
ax[0, 1].imshow(imgZit1);
ax[0, 1].set_title('Zit 1x Scale')
ax[1, 0].imshow(imgZit2);
ax[1, 0].set_title('Zit 2x Scale')
ax[1, 1].imshow(imgZit4);
ax[1, 1].set_title('Zit 4x Scale')
```

```python
# Load images as arrays
arrayZit0 = np.array(Image.open(('psnrEdit-1x.tif')))
arrayZit1 = np.array(Image.open(('psnrOrig.tif')))
arrayZit2 = np.array(Image.open(('psnrEdit-2x.tif')))
arrayZit4 = np.array(Image.open(('psnrEdit-4x.tif')))

# Calculate PSNR values
psnrZit0 = msr.compare_psnr(arrayZit0, arrayZit0)
psnrZit1 = msr.compare_psnr(arrayZit0, arrayZit1)
psnrZit2 = msr.compare_psnr(arrayZit0, arrayZit2)
psnrZit4 = msr.compare_psnr(arrayZit0, arrayZit4)

# Print statements
print(f"PSNR No Zit & No Zit: {round(psnrZit0, 2)}")
print(f"PSNR No Zit & Zit 1x: {round(psnrZit1, 2)}")
print(f"PSNR No Zit & Zit 2x: {round(psnrZit2, 2)}")
print(f"PSNR No Zit & Zit 4x: {round(psnrZit4, 2)}")
```
